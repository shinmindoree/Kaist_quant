```python
import pandas as pd
import statistics
import numpy as np

data = pd.read_excel('./processed_quant_mid.xlsx')
data.set_index('Date', inplace=True)

#data의 3개월 rolling 수익률을 구하고 별도의 파일로 저장
rolling_returns_3m = data.pct_change(periods=3).iloc[:, 0:3684]
rolling_returns_6m = data.pct_change(periods=6).iloc[:, 0:3684]
rolling_returns_9m = data.pct_change(periods=9).iloc[:, 0:3684]
rolling_returns_12m = data.pct_change(periods=12).iloc[:, 0:3684]


```

```python
# 3x3
# 각 행별로 상위 30%, 하위 30% 값을 추출하는 컨디션
buy = rolling_returns_3m.apply(lambda x: x >= x.quantile(0.7), axis=1)
sell = rolling_returns_3m.apply(lambda x: x <= x.quantile(0.3), axis=1)


buy
```

```python
#위에서 추출한 종목을 3개월간 보유했을때 수익률을 각 행별로 구하고, 이들의 평균을 구한다

selected_stocks_returns = []
first_valid_indices = rolling_returns_3m.apply(lambda x: x.first_valid_index())

selected_stocks_returns = []
for column in rolling_returns_3m.columns:
    start_index = first_valid_indices[column]  # 첫 번째 수치가 나오는 행의 인덱스
    if start_index is not None:  # NaN 값이 없는 경우에만 수익률 계산
        selected_returns = rolling_returns_3m.loc[start_index:, column].rolling(window=3).mean()
        # NaN 값을 -1로 대체하여 계산
        selected_returns = selected_returns.fillna(-1)
        selected_stocks_returns.append(selected_returns.mean())

average_return = sum(selected_stocks_returns) / len(selected_stocks_returns)

print(selected_stocks_returns)
print(average_return)
np.nanmean(selected_stocks_returns)
```
